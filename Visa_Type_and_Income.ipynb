```python
#One of the things we wanted to explore in this data set was the relationship between visa types and income
#We wanted to see if the ways of getting a residence and work permit in US was connected to Income


#Lets import our data to begin our analysis:

import pandas as pd
import matplotlib.pyplot as plt
salaries_df = pd.DataFrame(pd.read_excel("salary_data_states_corrected.xlsx"))

#We know that our data has a column called 'Visa Class'. Lets use this property to group our data frame.

visa_sort_df = salaries_df.groupby('VISA_CLASS').mean().sort_values(by="PAID_WAGE_PER_YEAR", ascending=[False])
visa_sort_df = visa_sort_df.reset_index()
```

```python
#Lets also create a plot to have a visual perspective:
visa_sort_df.head().plot(x ='VISA_CLASS', y='PAID_WAGE_PER_YEAR', kind = 'bar', 
                     ylabel="USD", xlabel="Visa Type",    label="Paid Wage", color="g").set_title("Anual Incomes of Different Visa Types")
plt.show()
```

```python
salaries_df["VISA_CLASS"].value_counts()
```

```python
#Lets also check the distribution of jobs and visa types to have a better perspective:

E3_salaries_df = salaries_df.loc[salaries_df['VISA_CLASS'].isin(["E-3 Australian"])]
greencard_salaries_df = salaries_df.loc[salaries_df['VISA_CLASS'].isin(["greencard"])]
H1B1C_salaries_df = salaries_df.loc[salaries_df['VISA_CLASS'].isin(["H-1B1 Chile"])]
H1B1S_salaries_df = salaries_df.loc[salaries_df['VISA_CLASS'].isin(["H-1B1 Singapore"])]
H1B_salaries_df = salaries_df.loc[salaries_df['VISA_CLASS'].isin(["H-1B"])]
```

```python
E3_jobsort = E3_salaries_df.groupby('JOB_TITLE_SUBGROUP').count().sort_values(by="CASE_NUMBER", ascending=[False]).reset_index()
gc_jobsort = greencard_salaries_df.groupby('JOB_TITLE_SUBGROUP').count().sort_values(by="CASE_NUMBER", ascending=[False]).reset_index()
H1B1S_jobsort = H1B1S_salaries_df.groupby('JOB_TITLE_SUBGROUP').count().sort_values(by="CASE_NUMBER", ascending=[False]).reset_index()
H1B1C_jobsort = H1B1C_salaries_df.groupby('JOB_TITLE_SUBGROUP').count().sort_values(by="CASE_NUMBER", ascending=[False]).reset_index()
H1B_jobsort = H1B_salaries_df.groupby('JOB_TITLE_SUBGROUP').count().sort_values(by="CASE_NUMBER", ascending=[False]).reset_index()
```

```python
#For E3
E3_jobsort.plot(x ='JOB_TITLE_SUBGROUP', y='CASE_NUMBER', xlabel="Job Type", ylabel="Count",  kind = 'bar', label="Count").set_title("E-3 Au Job Distribution")
plt.show()
  
#For greencard
gc_jobsort.plot(x ='JOB_TITLE_SUBGROUP', y='CASE_NUMBER', xlabel="Job Type", ylabel="Count",  kind = 'bar', label="Count").set_title("Greencard Job Distribution")
plt.show()

#For H-1B1S
H1B1S_jobsort.plot(x ='JOB_TITLE_SUBGROUP', y='CASE_NUMBER', xlabel="Job Type", ylabel="Count",  kind = 'bar', label="Count").set_title("H-1B1S Au Job Distribution")
plt.show()
  
# For H-1B
H1B_jobsort.plot(x ='JOB_TITLE_SUBGROUP', y='CASE_NUMBER', xlabel="Job Type", ylabel="Count",  kind = 'bar', label="Count").set_title("H-1B Au Job Distribution")
plt.show()

# For H-1B1C
H1B1C_jobsort.plot(x ='JOB_TITLE_SUBGROUP', y='CASE_NUMBER', xlabel="Job Type", ylabel="Count",  kind = 'bar', label="Count").set_title("H-1B Au Job Distribution")
plt.show()
```
