```python
#One of the things we wanted to explore in this data set was the relationship between visa types and income
#We wanted to see if the ways of getting a residence and work permit in US was connected to Income


#Lets import our data to begin our analysis:

import pandas as pd
import matplotlib.pyplot as plt
salaries_df = pd.DataFrame(pd.read_excel("salary_data_states_corrected.xlsx"))

#We know that our data has a column called 'Visa Class'. Lets use this property to group our data frame.

visa_sort_df = salaries_df.groupby('VISA_CLASS').mean().sort_values(by="PAID_WAGE_PER_YEAR", ascending=[False])
visa_sort_df = visa_sort_df.reset_index()
```

```python
#Lets also create a plot to have a visual perspective:
visa_sort_df.head().plot(x ='VISA_CLASS', y='PAID_WAGE_PER_YEAR'
                                            , kind = 'bar').set_title("Anual Incomes of Different Visa Types")
```
